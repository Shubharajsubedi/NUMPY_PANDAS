```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
plt.style.use('default')
plt.rcParams['figure.figsize']=(6,4)
```

```python
x =np.linspace(0,100,6)
print(x)

y= np.sin(x)
print(y)
plt.plot(x,y)
plt.title("line plot - size wave")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```
